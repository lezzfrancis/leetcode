## Contains Duplicate

```python
def containsDuplicate(nums):
    if len(nums) > len(set(nums)):
        return True
    else:
        return False
        
```

```python
containsDuplicate([1,2,3,3])
```

```python
containsDuplicate([1,2,3,4])
```

## Valid Anagram

```python
def validAnagram(s,t):
    if sorted(s) == sorted(t):
        return True
    else:
        return False
        
```

```python
validAnagram("racecar", "carrace")
```

### Non Sorted Version

```python
from collections import defaultdict
def validAnagram2(s,t):
    
    d1, d2 = defaultdict(int), defaultdict(int)

    for c in s:
        d1[c] += 1

    for c in t:
        d2[c] += 1

    if d1 == d2:
        return True
    else:
        return False
        
    
        
```

```python
validAnagram2("racecar", "carrace")
```

```python
validAnagram2("car", "rat")
```

## Two Sum

```python
def twoSum(nums, target):

    d1 = {}
    for i,n in enumerate(nums):
        d1[n] = i
    for i,n in enumerate(nums):
        if target - n in d1:
            if d1[target-n] != i:
               return [i,d1[target-n]]
           

        
```

```python
twoSum([3,2,4],6)
```

## Group Anagram

```python
from collections import defaultdict
def groupAnagram(strs):

    d1 = defaultdict(list)
    for s in strs:
        d1[''.join(sorted(s))].append(s)

    return d1.values()
```

## Top K Frequent Elements

```python
def topk(nums,t):

    d = defaultdict(int)
    for n in nums:
        d[n]+=1

    items = [k for k,v in sorted(d.items(),key=lambda x: x[0]) if v >= t]
    return items

    print(d)
    

        
```
